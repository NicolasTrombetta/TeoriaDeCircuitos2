Para hallar la transferencia $H(s)=\frac{V_2}{V_1}$ del circuito planteamos nodos:
\begin{equation}\tag{1}
V_A (G_1+G_2)-V_1 G_1-V_2 G_2 = 0 
\end{equation}

\begin{equation}\tag{2}
V_A (S C_1+G_3)-V_1 S C_1 = 0
\end{equation}

Despejando $V_A$ de $(1)$ y $(2)$ e igualando obtenemos:
\begin{equation}
\frac{V_1 G_1+V_2 G_2}{(G_1+G_2)} = \frac{V_1 S C_1}{(S C_1+G_3)} 
\end{equation}

Finalmente obtenemos:
\begin{equation}
T(s)=\frac{V_2}{V_1}=\frac{S-\frac{R_2}{R_1 R_3 C_1}}{S+\frac{1}{R_3 C_1}}
\end{equation}

Si $R_2=R_1$ entonces el digrama de ceros y polos presenta simetría especular respecto del eje imaginario.





## Simulacion con sympy

```python
import sympy as sp
from pytc2.general import print_latex

# símbolos
V1, V2, VA= sp.symbols('V1 V2 VA')
R1, R2, R3, C1 = sp.symbols('R1 R2 R3 C1')
S=sp.symbols('S')

eqs=[
    VA*(1/R1+1/R2)-V1*(1/R1)-V2*(1/R2),
    VA*(S*C1+1/R3)-V1*S*C1
]

sol=sp.solve(eqs,[V1,V2,VA])

T=sol[V2]/sol[V1]
T=sp.simplify(T)

print_latex(T)
```
